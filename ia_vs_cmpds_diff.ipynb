```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('ia_vs_cmpds_diff.tsv', sep='\t')
diffs = ['1049-953','1150-1049','1250-1150','1350-1250','1450-1350','1550-1450','1650-1550','1750-1650','1852-1750','1950-1852','2050-1950','2150-2050','2250-2150','2294-2250']

for diff in diffs:
    g = sns.jointplot(data=data,
                        x=f"{diff}_cmpd_diff", 
                        y=f"{diff}_top10_diff",
                        kind="hex",
                        marginal_kws={'color': 'green', 'bins':30})
    
    x1 ,x2 = -3, 50
    y1, y2 = -6, 100
    g.ax_joint.set_xlim(x1, x2)
    g.ax_joint.set_ylim(y1, y2)
    g.ax_marg_x.set_xlim(x1, x2)
    g.ax_marg_y.set_ylim(y1, y2)
        
    plt.show()
```
