```python
import seaborn as sns
import pandas as pd

def create_jointplot(data, num_approved=None):
    data = data[data['num_approved'] == num_approved] if num_approved is not None else data
    x1, y1 = 0, 0
    x2, y2 = 50, 100
    g = sns.jointplot(data=data, 
                      x='num_cmpds', 
                      y='ia', 
                      hue='num_approved',
                      palette='colorblind', 
                      kind='hist', 
                      bins=(range(x1, x2+1, 1), range(y1, y2+1, 5)))
    g.ax_joint.set_xlim(x1, x2)
    g.ax_joint.set_ylim(y1, y2)
    title = f'num_approved = {num_approved}' if num_approved is not None else 'num_approved_hue'
    g.fig.suptitle(title, y=1.02)

    return g

data = pd.read_csv('jointplot_marginal_hist.tsv', sep='\t')
create_jointplot(data)
create_jointplot(data, 1250)
create_jointplot(data, 1350)
create_jointplot(data, 1450)
```
