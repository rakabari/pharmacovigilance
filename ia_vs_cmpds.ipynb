```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('ia_vs_cmpds.tsv', sep='\t')
cmpds = [953, 1049, 1150, 1250, 1350, 1450, 1550, 1650, 1750, 1852, 1950, 2050, 2150, 2250, 2294]

for num in cmpds:
    g = sns.jointplot(data=data,
                        x=f"{num}_cmpds", 
                        y=f"{num}_top10",
                        kind="hex",
                        marginal_kws={'color': 'green', 'bins':30})
    
    x1 ,x2 = -3, 50
    y1, y2 = -6, 100
    g.ax_joint.set_xlim(x1, x2)
    g.ax_joint.set_ylim(y1, y2)
    g.ax_marg_x.set_xlim(x1, x2)
    g.ax_marg_y.set_ylim(y1, y2)
        
    plt.show()
```
